```python
def func2(x):
    if x == 1:
        return 'Horrible'
    elif x <= 4:
        return 'OK'
    else:
        return 'Great'
```

```python
user_stars['rating2'] = user_stars['stars'].apply(func2)
user_stars.head()
```

```python
def rating_groups(df):
    if df['stars'] == 1:
        return 'Horrible'
    elif df['rating'] == 'Bad':
        return 'OK'
    else:
        return 'Fine'
```

```python
user_stars['rating3'] = user_stars.apply(rating_groups, axis=1)
user_stars.head()
```
