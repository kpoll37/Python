```python
def embark(df):
    if df['embarked'] == "S":
        return 'Southampton'
    elif df['embarked'] == "C":
        return 'Cherbourg'
    elif df['embarked'] == "Q":
        return 'Queenstown'
    else:
        return df['embarked']
```

```python
titanic1d[0:15]
```

```python
titanic1d['embarked'] = titanic1d.apply(embark, axis=1)
```

```python
titanic1d['fare'] = titanic1d['fare'].fillna(0)
titanic1d[0:25]
```

```python
titanic1d['Family_group']= titanic1d['sibsp']+titanic1d['parch']
```

```python
def family(df):
    if df['Family_group'] == 0:
        return 'Solo'
    elif df['Family_group'] < 4:
        return 'Small Family'
    else:
        return 'Family'
```

```python
titanic1d['Family_group']=titanic1d.apply(family, axis=1)
```

```python
feature_cols = ['Age_adj', 'fare', 'pclass_1', 'pclass_2',
       'pclass_3', 'gender_female', 'gender_male', 'embarked_Cherbourg',
       'embarked_Queenstown', 'embarked_Southampton', 'Title_ Master',
       'Title_ Miss', 'Title_ Mr', 'Title_ Mrs', 'Title_Other',
       'Family_group_Family', 'Family_group_Small Family',
       'Family_group_Solo']
```

```python
X=titanic3[feature_cols]
y=titanic3.survived
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=1)
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth=2,
                                  class_weight ='balanced')
```

```python
clf2 = clf.fit(X_train, y_train)
```

```python
tree.plot_tree?
```

```python
tree.plot_tree(clf2)
```

```python
tree.plot_tree(clf2,
    feature_names=feature_cols,
    class_names=['died', 'survived'],
    filled=True, precision=2,
    node_ids=True, fontsize=12,
    label = 'root', proportion=True,
    impurity=False)
```

```python
from sklearn import metrics
```

```python
y_pred= clf2.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
metrics.confusion_matrix(y_test,y_pred)
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(y_test, y_pred).ravel()
```

```python
precision = tp/ (tp+fp)
precision
```

```python
recall = tp/(tp+fn)
recall
```

```python
print(metrics.classification_report(y_test, y_pred))
```
